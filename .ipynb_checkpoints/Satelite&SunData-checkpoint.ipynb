# Satelite Data

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
NDVI_Modis = pd.read_csv('NDVI_Modis.csv',parse_dates=['Date'])
NDVI_Modis = NDVI_Modis.set_index(pd.DatetimeIndex(NDVI_Modis['Date']))
NDVI_Modis=NDVI_Modis.drop('Date',axis=1)
NDVI_Landsat7 = pd.read_csv('Landsat7_NDVI.csv',parse_dates=['Date'])
NDVI_Landsat7 = NDVI_Landsat7.set_index(pd.DatetimeIndex(NDVI_Landsat7['Date']))
NDVI_Landsat7=NDVI_Landsat7.drop('Date',axis=1)
NDVI_Landsat8 = pd.read_csv('Landsat8_NDVI.csv',parse_dates=['Date'])
NDVI_Landsat8 = NDVI_Landsat8.set_index(pd.DatetimeIndex(NDVI_Landsat8['Date']))
NDVI_Landsat8=NDVI_Landsat8.drop('Date',axis=1)
Snow = pd.read_csv('Snow_Cover.csv',parse_dates=['Date'])#NDSI_Snow_Cover
Snow = Snow.set_index(pd.DatetimeIndex(Snow['Date']))
Snow=Snow.drop('Date',axis=1)
```

```python
Data = NDVI_Modis.copy()
Data = Data.join(NDVI_Landsat7,how='outer')
Data = Data.join(NDVI_Landsat8,how='outer')
Data = Data.join(Snow,how='outer')
Data['DOY'] = Data.index.dayofyear

print(Data['DOY'])
Yearly = Data.groupby(Data.index.month).mean()

print(Data.head())
fig,ax=plt.subplots(figsize=(8,6))
ax.scatter(Data['NDVI_Landsat7'],Data['NDVI_MODIS'])
ax.set_xlabel('Landsat')
ax.set_ylabel('MODIS')
# print(Da)
```

```python
Data['SnowFree']=Data['NDSI_Snow_Cover'].interpolate().bfill()
Data['SnowFree_Intep'] = 0
Data.loc[np.isnan(Data['NDSI_Snow_Cover'])==True,'SnowFree_Interp']=1

# Data['SnowFree'].plot()
Data.loc[Data['SnowFree']<=20,'SnowFree']=1
Data.loc[Data['SnowFree']>20,'SnowFree']=-1
Data.loc[Data['SnowFree']<0,'SnowFree']=0
for Y in range(2008,2022):
    Temp = Data.loc[Data.index.year==Y]
    plt.plot(Temp.DOY,Temp['SnowFree'])
# print(Data['DOY'])
print(Data.resample('Y').sum()[{'SnowFree','SnowFree_Interp'}])
# print(Data['NDSI_Snow_Cover'].describe())
```

```python
import statsmodels.api as sm
import numpy as np
# duncan_prestige = sm.datasets.get_rdataset("Duncan", "carData")
Temp = Data.loc[Data['NDSI_Snow_Cover']<=0,['NDVI_MODIS','NDVI_Landsat7','DOY']].dropna()
X = Temp[['NDVI_MODIS','DOY']]
Y = Temp['NDVI_Landsat7']
X = sm.add_constant(X)
model = sm.OLS(Y,X)
results = model.fit()
results.params

fig = plt.figure()
plt.plot(Data['DOY']*results.params[2]+Data['NDVI_MODIS']*results.params[1]+results.params[0])
```

```python
print(results.summary())
```
